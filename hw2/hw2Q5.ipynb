```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")
print("The train columns are:")
li = train.columns.tolist()
for i in range(len(li)):
  print(li[i])

print("\n")

print("The test columns are:")
li2 = test.columns.tolist()
for i in range(len(li2)):
  print(li2[i])
```

```python
train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")

train_df = train.drop(columns=["Unnamed: 0", "zipcode"])
test_df  = test.drop(columns=["Unnamed: 0", "zipcode", "id", "date"])

train_df['price'] = train_df["price"] / 1000
test_df['price']  = test_df["price"] / 1000

X_train = train_df.drop(columns=["price"])
y_train = train_df["price"].values

X_test = test_df.drop(columns=["price"])
y_test = test_df["price"].values

```

```python
feat = X_train.columns.tolist()

scalar = StandardScaler()
scaled_X_train = scalar.fit_transform(X_train)
scaled_X_test  = scalar.transform(X_test)


ones_arr1 = np.ones((scaled_X_train.shape[0], 1))
X_train_final = np.concatenate([ones_arr1, scaled_X_train], axis=1)

ones_arr2 = np.ones((scaled_X_test.shape[0], 1))
X_test_final = np.concatenate([ones_arr2, scaled_X_test], axis=1)
```

```python
alphas = [0.01, 0.1, 0.5]
iters = [10, 50, 100]

li1 = []
li2 = []
li3 = []
li4 = []
li5 = []
li6 = []

n = X_train_final.shape[0]

```

```python
#part 1 and 2 gradient descent along with different learning rates and iteration

for a in alphas:
    for i in iters:
        theta = np.zeros(X_train_final.shape[1])
        for j in range(i):
            prediction = X_train_final @ theta
            err = prediction - y_train
            grad = (2 / n) * (X_train_final.T @ err)
            theta = theta - (a * grad)


        print(f"Alpha={a}, Iterations={i}")
        print(f"Theta={theta}")
        print("\n"*4)

        pred1 = X_train_final @ theta
        pred2  = X_test_final @ theta

        train_mse = mean_squared_error(y_train, pred1)
        train_r2  = r2_score(y_train, pred1)
        test_mse  = mean_squared_error(y_test, pred2)
        test_r2   = r2_score(y_test, pred2)

        li1.append(a)
        li2.append(i)
        li3.append(train_mse)
        li4.append(train_r2)
        li5.append(test_mse)
        li6.append(test_r2)

```

```python
df = pd.DataFrame({
    "Alpha": li1,
    "Iterations": li2,
    "Train MSE": li3,
    "Train R2": li4,
    "Test MSE": li5,
    "Test R2": li6
})
```

```python

print(df)
```

When the learning rate is small for example if it is α=0.01, then model converges slowly. You can see at 10 iterations its perfomance is bad. But it keeps on improving as it can seen at 50 iterations. But it still doesnt converge at 100 iterations.

Next if we increase learning rate a little to α=0.1, then model would converge faster. Model reaches R² = 0.70 within almost only 10 iterations and almost converges in 100 iteration


If learning rate is too big like here it is α=0.5 then model will diverge. During gradient descent model will overshoot the convergence point.

Here I believe the best learning rate is α=0.1 which almost converges around 100 iterations
