```python
import pandas as pd
import numpy as np

```

```python
# Load the dataset from the csv file
df = pd.read_csv('kc_house_data.csv')

```

```python
# Dropped columns id, date, and zipcode
dropColumns = ['id', 'date', 'zipcode']
new_df = df.copy()

for col in dropColumns:
  if col in new_df.columns:
    new_df.drop(col, axis = 1, inplace=True)

# It has seperated the features from the target variable
target = 'price'
feat = []
for col in new_df.columns:
  if col != target:
    feat.append(col)

print(feat)
```

```python
#part 1

# Initialize empty lists for statistics
averages = []
mins = []
maxs = []
variances = []

# Iterates each feature and finds out the avg, min,max, avriance
for f in feat:
    averages.append(new_df[f].mean())
    mins.append(new_df[f].min())
    maxs.append(new_df[f].max())
    variances.append(new_df[f].var())

# Create DataFrame with the collected statistics
stats_df = pd.DataFrame({
    'Feature': feat,
    'Average': averages,
    'Min': mins,
    'Max': maxs,
    'Variance': variances
})


print(stats_df.to_string(index=False))



min_avg_i = stats_df['Average'].idxmin()
max_avg_i = stats_df['Average'].idxmax()

print("\n"*3)

print(f"Feature with lowest average: {stats_df.loc[min_avg_i, 'Feature']} and value is {stats_df.loc[min_avg_i, 'Average']}")

print(f"Feature with highest average: {stats_df.loc[max_avg_i, 'Feature']} and value is {stats_df.loc[max_avg_i, 'Average']}")


min_var_idx = stats_df['Variance'].idxmin()
max_var_idx = stats_df['Variance'].idxmax()


print(f"Feature with lowest variance: {stats_df.loc[min_var_idx, 'Feature']} and variance is {stats_df.loc[min_var_idx, 'Variance']}")
print(f"Feature with highest variance: {stats_df.loc[max_var_idx, 'Feature']} and variance is {stats_df.loc[max_var_idx, 'Variance']}")
```

```python
#part 2



# Finds corelation of each feature with price
corrLi = []
for f in feat:
    corr = new_df[f].corr(new_df[target])
    corrLi.append(corr)


corr_df = pd.DataFrame({
    'Feature': feat,
    'Correlation with Price': corrLi
})


corr_df_s = corr_df.sort_values('Correlation with Price', ascending=False)

print("\nCorrelation Coefficients Table")
print()
print()
for i in range(len(corr_df_s)):
    feature = corr_df_s.iloc[i]['Feature']
    correlation = corr_df_s.iloc[i]['Correlation with Price']
    print(f"{feature}: {correlation}")

# Finds which feature is positively correlated features
positive_corr = corr_df[corr_df['Correlation with Price'] > 0]

print()
print()
print()



print(f"Number of positively correlated features: {len(positive_corr)}")
print("\nFeatures with positive correlation:")
s_positive = positive_corr.sort_values('Correlation with Price', ascending=False)

for i in range(len(s_positive)):
    feature = s_positive.iloc[i]['Feature']
    correlation = s_positive.iloc[i]['Correlation with Price']
    print(f"  {feature}: {correlation}")


max_corr_idx = corr_df['Correlation with Price'].idxmax()
print()
print()
print(f"Feature with highest positive correlation is {corr_df.loc[max_corr_idx, 'Feature']} and value is {corr_df.loc[max_corr_idx, 'Correlation with Price']}")


```

```python
#part 3


#finds features with negative correlation
negative_corr = corr_df[corr_df['Correlation with Price'] < 0]

if len(negative_corr) > 0:
    print(f"There are {len(negative_corr)} feature(s) found with negative correlation:")
    sorted_df = negative_corr.sort_values('Correlation with Price')
    for i in range(len(sorted_df)):
      negFeat = sorted_df.iloc[i]['Feature']
      negCorr = sorted_df.iloc[i]['Correlation with Price']
      print(f"  {negFeat}: {negCorr}")
else:
    print("\nNo features with negative correlation found.")

```
